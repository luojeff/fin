# Futures Markets

A buyer holds a *long futures position*. A seller holds a *short futures position*.

Most positions do not lead to delivery, as traders can close out their positions prior to delivery period. The gain or loss is determined by **the change in the futures price between the first trade and the closing out transaction**.

### Specification of Contract

**Underlying Asset**

For commodities, the exchange determines what grade(s) of the commodity are acceptable.

For Treasury notes, the exchange adjusts the price depending on the *coupon* and *maturity date* of the bond.

**Contract size**

Specifies the amount of the asset that has to be delivered under one contract. Exchanges must find the right balance:
- If the contract size is too small, transactional cost is too high
- If it's too high, many small investors may not be willing to participate

Some exchanges have introduced *mini contracts* to attract smaller investors.

**Location and time of delivery**

Futures contracts are referred to by their delivery month. For many contracts, the delivery period is *the whole month*. 

Trading typically ends a few days before the last day on which the delivery can be made.

### Price Quotation

Exchanges define how prices are quoted. 

Certain futures are quoted in dollars and cents. Treasury bonds and notes are quoted in dollars and *thirty-seconds* of a dollar.

### Price Movement Limits / Circuit Breakers

Exchanges impose restrictions on how much prices can move daily.

A maximum move up and down are called *limit up* and *limit down* movements. When these reach a certain magnitude, trading may cease for the day. 

There is controversy on whether this is a good thing. 

- 7%: Market trading halt
- 13%: Market trading halt
- 20%: Cash equity market and US-equity index and options cease trading for the day

### Position Limits

Defines the maximum number of contracts a speculator may hold.

### Convergence

As you approach the delivery date, the futures price converges to the spot price. The futures price can be above or below the spot price at any point during this convergence. 

If this wasn't the case, there is an arbitrage opportunity (buy future, sell underlying asset, or vice versa). 

### Margin Accounts

To ensure that buyers and sellers can fulfill their end, a *margin account* account is operated between these participants through a broker. 

Daily changes in futures prices are reflected in these margin prices. When the futures value drops below the *maintenance margin* threshold (typically 75% of initial margin), a *margin call* is submitted to the investor requiring the deposit of additional funds called the *variation margin*. If no funds are deposited, the position is closed. 

**Minimum margin requirements are set by the exchange clearing house by determining the variability of the underlying asset's price**, though individual brokers can require more. 

Through *mark-to-market*, a futures contract is **settled daily**, brining the value of the contract back to zero. 

Initial margin requirements may be satisfied in Treasury bills or stock at a portion of their value.

The objectives of the trader also determine margin requirements:
- Hedger
- Day trader
- Spread transactions

Unlike the spot market, margin requirements on long and short positions are the same due to symmetry.

### Clearing House

Intermediary in futures transaction. 

Brokers who aren't members must channel their business and post margins with the member of the clearinghouse. 

Initial margin is often determined on a net basis. 

### Central Counterparty (CCP)

Clearing houses for standard OTC transactions. Counterparty to both participants in an OTC trade.

The counterparty takes on the credit risk of both A and B, as it:
1. Buys asset from B at the future period for the agreed price
2. Sells the asset to A at the future period for the agreed price

CCP members must post initial margins to CCP. Transactions are valued daily. If needed, the CCP requests variation margin payments from the member. Analogous to the relationship between broker and clearinghouse member.

### Bilateral Clearing

Direct master agreement between two parties using a collateral agreement called the *credit support annex* (CSA).

e.g. When a transaction between A and B increases by $X$ for A, B must provide collateral worth $X$ (equivalent to the variation margin) to A.

Long-Term Capital Management (LTCM) bilaterally cleared derivatives, allowing them to be highly levered. 

Note that transactions are not settled daily in OTC markets, so daily variation margin earns interest just like initial margin.

### Quotation

Three prices are typically quoted:
1. Open price
2. Highest daily price
3. Lowest daily price

The *settlement price* determines the daily gain and loss for margin requirements. Usually calculated as the **trade price of the contract immediately before the end of the day's trading session**.

The *open interest* denotes the number of contracts (long or short) outstanding.

The *trading volume* denotes the number of contracts traded in a day.

### Futures Market Patterns

Normal market: Settlement futures prices increase with maturity (of the contract).

Inverted market: Settlement futures prices decline with maturity.

(Also referred to as *contango* and *backwardation* respectively)

### Delivery

Very few contracts actually lead to delivery, as most are closed out prior. However, the possibility of eventual delivery determines the futures price.

Because contracts can be traded many times, the original participants in a contract may not be the same at maturity.

The exchange typically passes the *notice of intention to deliver* onto the party with the oldest outstanding long position. Long position parties must accept delivery notices. These notices may be transferrable within a short period of time. 

A contract has three critical days:
1. First notice day - *Notice of intention to deliver* is submitted to the exchange
2. Last trading day
3. Last notice day

**Buyer should close contracts prior to first notice day** to avoid having to take delivery.

If the underlying asset is not a commodity or tangible good, a contract may be *settled in cash*, where all outstanding constracts are declared closed on a predetermined day at a certain settlement price.

### Traders

Can be categorized as:
1. Futures Commission Merchants (FCMs)
2. Locals

Can also be categorized as:
1. Hedgers
2. Speculators
    - Scalpers: Hope to profit from short-term changes in contract price
    - Day traders: Hold their positions for less than one trading day, and refuse to take the risk of overnight adverse news
    - Position traders: Holding period is much longer, as they hope to profit from major movements in the markets
3. Arbitrageurs

### Orders

1. Market order
    - Request that trade be carried out immediately at best price in market
2. Limit order
    - Order can be executed at a specific price or one more favorable
    - No guarantee an order will be executed
3. Stop-loss order
    - *Becomes a market order* once the specified price (or one more unfavorable) is hit
    - Can close out a position if **unfavorable price movements occur**
    - Might still give you bad a price in illiquid markets
4. Stop-limit order
    - *Becomes a limit order* once bid or offer is at a price equal to or less favorable than (at or through) the stop price
    - **Stop price and limit price** must be specified
5. Market-if-touched order
    - *Becomes a market order* once specific price is hit
    - Similar to *stop order*, though it secures profits if **favorable price movements occur**, instead of mitigating losses
6. Discretionary / Market-not-held order
    - Trades as a market order, though may be delayed by broker attempting to get a better price


Time conditions:
1. Time-of-day
    - Specifies a particular period of time during day when the order can be executed
2. Open / Good-till-cancelled order
    - In effect until executed or until end of trading in the particular contract
3. Fill-or-kill
    - Must be executed immediately in its entirety, or not at all
